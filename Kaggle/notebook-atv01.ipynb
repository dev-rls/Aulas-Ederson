```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/kaggle/input/sorveteria/Untitled spreadsheet - flavors.csv')

# Sabores mais vendidos
sabores = df['flavor'].value_counts().head(5)
sabores.plot(kind='bar', color='grey')

for i, v in enumerate(sabores):
    plt.text(i, v + 0.5, str(v), ha='center', va='bottom')

plt.title('Sabores Mais Vendidos')
plt.xlabel('Sabor')
plt.ylabel('Quantidade')
plt.show()

# Sabores menos vendidos
menos_vend = df['flavor'].value_counts().tail(5)
menos_vend.plot(kind='bar', color='green')

for i, v in enumerate(menos_vend):
    plt.text(i, v + 0.5, str(v), ha='center', va='bottom')

plt.title('Sabores Menos Vendidos')
plt.xlabel('Sabor')
plt.ylabel('Vendidos')
plt.show()

# Quantidade vendidos por semana
semana = df['week'].value_counts().head(5)
semana.plot(kind='bar', color='orange')

for i, v in enumerate(semana):
    plt.text(i, v + 0.5, str(v), ha='center', va='bottom')

plt.title('Quantidade Vendidos por Semana')
plt.xlabel('Sabor')
plt.ylabel('Semana')
plt.show()

# Vendas por sabor
sabor = df['units sold'].value_counts().head(5)
sabor.plot(kind='bar', color='red')

for i, v in enumerate(sabor):
    plt.text(i, v + 0.5, str(v), ha='center', va='bottom')

plt.title('Vendas por Sabor')
plt.xlabel('Sabor')
plt.ylabel('Vendas')
plt.show()

# Campeão em vendas na semana 01
semana1 = df['week'].value_counts().head(5)
semana1.plot(kind='bar', color='yellow')

for i, v in enumerate(semana1):
    plt.text(i, v + 0.5, str(v), ha='center', va='bottom')

plt.title('Campeão em Vendas Semana 01')
plt.xlabel('Sabor')
plt.ylabel('Semana 01')
plt.show()

# Campeão em vendas mensal
mensal = df['units sold'].value_counts().head(5)
mensal.plot(kind='bar', color='grey')

for i, v in enumerate(mensal):
    plt.text(i, v + 0.5, str(v), ha='center', va='bottom')

plt.title('Campeão em Vendas Mensal')
plt.xlabel('Sabor')
plt.ylabel('Mensal')
plt.show()

# Campeão em vendas anual
anual = df['units sold'].value_counts().head(3)
anual.plot(kind='bar', color='purple')

for i, v in enumerate(anual):
    plt.text(i, v + 0.5, str(v), ha='center', va='bottom')

plt.title('Campeão em Vendas Anual')
plt.xlabel('Sabor')
plt.ylabel('Anual')
plt.show()

print(df.head(150))

```
